```python
def readfile():
    mem = []
    new_mem = []
    with open("input.txt",'r') as content:
        for line in content:
            mem.append(list(line.split()[0]))
    for i in range(len(mem)+6):
        new_list = []
        for j in range(len(mem[0])+6):
            if i<=2:
                new_list.append(".")
            elif i<= len(mem) -1 and j<=2:
                new_list.append(".")
            elif i<= len(mem) +2 and 3<=j<=len(mem[0])+2:
                new_list.append(mem[i-3][j-3])
            else:
                new_list.append(".")
        new_mem.append(new_list)
    return new_mem

def x_search(nestlist):
    x_coord = []
    xmas_counter = 0
    for i in range(len(nestlist)):
        for j in range(len(nestlist[0])):
            if nestlist[i][j] == "X":
                x_coord.append([i,j])
    for y,x in x_coord:
        pot_m_coord = [[y,x-1],[y,x+1],[y-1,x-1],[y-1,x],[y+1,x],[y+1,x+1],[y+1,x-1],[y-1,x+1]]
        del_list = []
        refined_m_coord = []
        for i in range(len(pot_m_coord)):
            y_y = pot_m_coord[i][0]
            y_x = pot_m_coord[i][1]
            if nestlist[y_y][y_x] != "M":
                del_list.append(i)
        for i in range(len(pot_m_coord)):
            if i not in del_list:
                refined_m_coord.append(pot_m_coord[i])
        for coords in refined_m_coord:
            pot_a = nestlist[coords[0]+(coords[0]-y)][coords[1]+(coords[1]-x)]
            pot_s = nestlist[coords[0]+2*(coords[0]-y)][coords[1]+2*(coords[1]-x)]
            if pot_a == "A" and pot_s == "S":
                xmas_counter +=1
    return xmas_counter

def a_search(nestlist):
    a_coord = []
    x_mas_counter = 0
    for i in range(len(nestlist)):
        for j in range(len(nestlist[0])):
            if nestlist[i][j] == "A":
                a_coord.append([i,j])
    for y,x in a_coord:
        diag_r = [[y-1,x-1],[y+1,x+1]]
        diag_l = [[y+1,x-1],[y-1,x+1]]
        top_r = nestlist[diag_l[1][0]][diag_l[1][1]]
        bot_r = nestlist[diag_r[1][0]][diag_r[1][1]]
        top_l = nestlist[diag_r[0][0]][diag_r[0][1]]
        bot_l = nestlist[diag_l[0][0]][diag_l[0][1]]
        if top_r ==  "M" and bot_l =="S" or top_r == "S" and bot_l == "M":
            if top_l == "M" and bot_r == "S" or top_l == "S" and bot_r == "M":
                x_mas_counter += 1
    return x_mas_counter

        

        
def main():
    data = readfile()
    print(x_search(data))
    print(a_search(data))

main()
```
