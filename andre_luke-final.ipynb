```python
def readfile():
    reports = {}
    counter = 0
    with open("input.txt",'r') as content:
        for line in content:
            split = line.split()
            for i in range(len(split)):
                split[i] = int(split[i])
            reports[counter] = split
            counter += 1
    return reports

def safe_test(dic):
    unsafe = 0
    for keys in dic:
        dec = False
        inc = False

        if 0 < dic[keys][0] - dic[keys][1] <= 3:
            dec = True
        elif 0 < dic[keys][1] - dic[keys][0] <= 3:
            inc = True
        else:
            if list_test_damp(dic[keys]) == False:
                unsafe +=1

        if inc == True:
            for i in range(len(dic[keys])-1):
                if 0 < dic[keys][i+1] - dic[keys][i] <= 3:
                    pass
                else:
                    if list_test_damp(dic[keys]) == False:
                        unsafe += 1
                        break
        elif dec == True:
            for i in range(len(dic[keys])-1):
                if 0 < dic[keys][i] - dic[keys][i+1] <= 3:
                    pass
                else:
                    if list_test_damp(dic[keys]) == False:
                        unsafe +=1
                        break
    safe = len(dic) - unsafe
    return safe

def list_test_damp(levels):
    #return False #remove # for part 1
    #[52, 51, 54, 57, 59, 62, 64] ruined my first solution
    #which birthed the mass_dampener
    alt_lists = mass_dampener(levels)
    
    for keys in alt_lists:
        dec = False
        inc = False
        alt_1_bad = False
        alt_1 = alt_lists[keys]
        if 0 < alt_1[0] - alt_1[1] <= 3:
            dec = True
        elif 0 < alt_1[1] - alt_1[0] <= 3:
            inc = True
        else:
            alt_1_bad = True
        
        if inc == True:
            for i in range(len(alt_1)-1):
                if 0 < alt_1[i+1] - alt_1[i] <= 3:
                    pass
                else:
                    alt_1_bad = True
                    break

        elif dec == True:
            for i in range(len(alt_1)-1):
                if 0 < alt_1[i] - alt_1[i+1] <= 3:
                    pass
                else:
                    alt_1_bad = True
                    break
        if alt_1_bad == False:
            return True
    return False
                
    
def mass_dampener(bad_list):
    new_lists = {}
    for i in range(len(bad_list)):
        new_lists[i] = bad_list.copy()
        new_lists[i].pop(i)
    return new_lists
        


def main():
    #print(readfile())
    print(safe_test(readfile()))

main()
```
